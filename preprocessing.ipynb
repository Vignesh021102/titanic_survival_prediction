```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
sns.set()
```

```python
random_state = 20
np.random.random_state =  random_state
```

```python
raw_df = pd.read_csv('./train.csv')
```

```python
raw_df['Embarked'] = raw_df['Embarked'].fillna('')
np.sort(raw_df['Embarked'].unique())
```

```python
np.sort(raw_df['Ticket'].unique())
```

```python
df = raw_df.copy()
```

```python
df['Sex'] = np.where(df['Sex'] == 'male',1,0)
```

```python
df = df.drop(['Name','Ticket'],axis=1)
```

```python
df['Cabin'] = df['Cabin'].fillna('000')
df['Embarked'] = df['Embarked'].fillna('000')
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
PclassGroup = pd.get_dummies(df['Pclass'],drop_first=True)
PclassGroup.columns = ['class2','class3']
```

```python
embarked = pd.get_dummies(df['Embarked'],drop_first=True)
embarked.columns = ['cherbourg','queenstown','southampton']
```

```python
cabin = pd.get_dummies(df['Cabin'].apply(lambda x:x[0]),drop_first=True)
cabin.columns = ['cabinA', 'cabinB', 'cabinC', 'cabinD', 'cabinE', 'cabinF', 'cabinG', 'cabinT']
```

```python
df = df.drop(['PassengerId','Pclass','Embarked','Cabin'],axis=1)
```

```python
preprocessed_df = pd.concat([df,PclassGroup,cabin,embarked],axis=1)
#preprocessed_df.to_csv('preprocessed_df.csv')
```

```python
survivedCount = preprocessed_df.groupby(['Survived']).count().values[:,0]
```

```python
survivedCount
```

```python
shuffled_predf = preprocessed_df.sample(frac = 1,random_state = random_state)
```

```python
zeros = 0
ones = 0
indexBal = 0
for i in range(shuffled_predf.shape[0]):
    if shuffled_predf.iloc[i+indexBal]['Survived'] == 0:
        zeros += 1
        if (zeros>survivedCount[1]):
            shuffled_predf = shuffled_predf.drop(shuffled_predf.iloc[i+indexBal].name)
            indexBal -= 1
    #print(i,zeros,ones,shuffled_predf['Survived'].iloc[i+indexBal],indexBal)
```

```python
shuffled_predf.groupby(['Survived']).count().values[:,0]
balanced_df = shuffled_predf
balanced_df.reset_index(drop=True)
```

```python
balanced_df.to_csv('preprocessed_df.csv')
```
